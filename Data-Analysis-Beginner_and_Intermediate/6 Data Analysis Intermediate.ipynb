```python
#HOW DO I FILTER ROWS OF A PANDAS DATAFRAME BY COLUMN VALUE

booleans = []
for i in movie.duration:
    if i >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[:5]
```

```python
len(booleans)
```

```python
#converting boolean columns to Series
s = pd.Series(booleans)
```

```python
s.head()
```

```python
print(s[3:7])
```

```python
#drop the False value by default
#it will show the full dataset
movie[s]
```

```python
movie[s].shape
```

```python
(s==True).value_counts()
```

```python
#HOW TDO I APPLY MULTIPLE FILTER CRITERIA TO A PANDAS DATAFRAME

movie = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movie.shape
```

```python
movie.head()
```

```python
movie.loc[(movie.duration >= 200)]
```

```python
movie.loc[(movie.duration >= 200)].shape
```

```python
#will print only boolean value

[(movie.duration >=200) & (movie.genre == 'Drama')]
```

```python
[(movie.duration >= 200) | (movie.genre == 'Drama')]
```

```python
movie.loc[(movie.duration >=200) | (movie.genre == 'Drama')]
```

```python
#multiple column filter
movie.loc[(movie.duration >= 200) & (movie.genre == 'Drama')]
```

```python
movie.loc[(movie.duration >=200) & (movie.genre == 'Drama') & (movie.title == 'Hamlet')]
```

```python
movie.loc[(movie.star_rating == 7.8) | (movie.title =='Hamlet') | (movie.genre == 'Drama')]
```

```python
#printing three type of movie
movie.loc[(movie.genre=='Drama') | (movie.genre=='Action') | (movie.genre=='Crime')]
```

```python
#shortly writing upper line of the code
movie.loc[movie.genre.isin(['Drama','Action','Crime'])]
```
