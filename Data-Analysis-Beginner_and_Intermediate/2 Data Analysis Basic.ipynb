```python
#constructing dataframe

area_dict = {'sylhet':100, 'dhaka':200, 'rajshahi':300, 'barisal':400, 'maymansingh':500, 'chittaganj':600}
s = pd.Series(area_dict)
df = pd.DataFrame(s, columns=['population'])
print(df)
#from a list
data = [{'a':i, 'b':i*2} for i in range(3)]
df = pd.DataFrame(data)
print(df)
df = pd.DataFrame([{'a':1, 'b':2}, {'b':3, 'c':4}])
print(df)
#from a dictionary
area_dict = {'sylhet':100, 'dhaka':200, 'rajshahi':300, 'barisal':400, 'maymansingh':500, 'chittaganj':600}
s = pd.Series(area_dict)
population_dict = {'sylhet':100, 'barisal':200, 'dhaka':300, 'rajshahi':400, 'chittaganj':500, 'mymansingh':600}
df = pd.DataFrame({'population':population_dict, 'area':s})
print(df)
#from a two dimentional numpy array
df = pd.DataFrame(np.random.rand(3,2), index=['a','b','c'], columns=['A','B'])
print(df)
```

```python
#pandas index object

ind = pd.Index([2,4,6,1,3,5])
print(ind)
print(ind[2])
print(ind[::2])
print(ind[1:4])
#index as orderset
indA = pd.Index([1,3,5,7,9])
indB = pd.Index([2,3,6,7,10])
print(indA & indB)
print(indA | indB)
print(indA ^ indB)
```

```python
#use of loc, iloc, ix 
dict = {'A':[3,4,6,8,1,3,5,7,9], 'B':[1,3,5,7,9,2,4,6,8]}
df = pd.DataFrame(dict, index=['a','b','c','d','e','f','g','h','i'])
print(df)

a_loc = df.loc['g']
print(a_loc)
b_loc = df.loc['c':'g']
print(b_loc)
c_loc = df.loc['d':'g','A':'B']
print(c_loc)

print('after a big break \n\n\n\n\n')
a_iloc=df.iloc[3]
print(a_iloc)
b_iloc=df.iloc[3:7]
print(b_iloc)
c_iloc=df.iloc[2:6, 0:2]
print(c_iloc)

print('after a big break\n\n\n\n')
a_ix=df.ix[3:8, 'A':'B']
print(a_ix)
```

```python
#dictionary in DataFrame

state = pd.Series({'sylhet':100, 'dhaka':200, 'chittaganj':300, 'rajshahi':400, 'barisal':500, 'maymansingh':600})
country = pd.Series({'chittaganj':300, 'rajshahi':100, 'dhaka':400, 'maymansingh':500, 'barisal':200, 'sylhet':600})
df = pd.DataFrame({'state':state, 'country':country})
print(df)
print(df.state)
print(df.state is df['state'])
print(df.country is df['country'])
print('after a big break\n\n\n\n')
#adding a new column in the DataFrame
df['distance'] = df['state']/ df['country']
print(df)
#DataFrame is converting to 2D array
print('after a big break\n\n\n\n')
print(df.values)
print(df.values[1])
print(df.T)
#use of ix in DataFrame
print('after a big break\n\n\n\n')
print(df)
print(df.ix['chittaganj':'rajshahi', 0:2])
#use of numpy-data access pattern
b = df[df.distance>2]
print(b)
a = df.loc[df.distance>=2, ['state','country']]
print(a)

#replaceing data of DataFrame
print('after a big break\n\n\n\n')
df.iloc[2,1] = 10000
print(df)
```
