# Resampling

```python
#every week's mean value
columns = ['Consumption', 'Wind','Solar', 'Wind+Solar']
weekly_mean = data[columns].resample('W').mean()
weekly_mean
```

```python
print('Shape of the original dataset :',data.shape)
print('Shape of the weekenly mean dataset :', weekly_mean.shape)
```

```python
#six month's daily and weekly consumption
fig, ax = plt.subplots()
ax.plot(data.loc['2017-01':'2017-06', 'Solar'], marker='.', linestyle='-', linewidth=.5, label='Daily consumption')
ax.plot(weekly_mean.loc['2017-01':'2017-06', 'Solar'], marker='o', linestyle='-', linewidth=.5, label='weekly consumption')
plt.xlabel('Date')
plt.ylabel('value')
plt.title('Daily and Weekly data')
plt.legend(loc='best')
```

```python
#every months mean value
columns = ['Consumption', 'Wind','Solar', 'Wind+Solar']
monthly_mean = data[columns].resample('M').mean()
monthly_mean
```

```python
#six month's monthly and weekly consumption
fig, ax = plt.subplots()

ax.plot(weekly_mean.loc['2017-01':'2017-06', 'Consumption'], marker='.', linestyle='-', linewidth=.5, label='weekly consumption')
ax.plot(monthly_mean.loc['2017-01':'2017-06', 'Consumption'], marker='o', linestyle='-', linewidth=.5, label='Monthly consumption')
plt.xlabel('')
plt.ylabel('')
plt.title('')
plt.legend(loc='best')
```

```python
columns = ['Consumption','Wind','Solar', 'Wind+Solar']
monthly_sum = data[columns].resample('M').sum()
monthly_sum
```

```python
#sum of consumption monthly
fig, axes = plt.subplots()
axes.plot(monthly_sum['Consumption'], label='monthly consumption', color='black')
monthly_sum[['Wind','Solar']].plot.area(ax=axes, linewidth=0)
ax.xaxis.set_major_locator(mdates.YearLocator())
```

```python
#sum in yearly consumption
columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
yearly_sum = data[columns].resample('Y').sum()
yearly_sum = yearly_sum.set_index(yearly_sum.index.year)
yearly_sum
```

```python
yearly_sum['Consumption'].plot(kind='bar')
plt.xticks(rotation=0)
```

```python
#yearly consumption figure
fig, ax = plt.subplots()
ax.plot(yearly_sum['Consumption'], color='yellow', linewidth=5)
```

```python
yearly_sum['Wind+Solar'].plot(kind='bar')
plt.xlabel('year')
plt.ylabel('total consumption')
plt.title('Wind and Solar consumption in every year')
```

```python
#wind and solar consumption wher have data
yearly_sum.loc['2012':,'Wind+Solar'].plot(kind='bar')
plt.xlabel('year')
plt.xticks(rotation=0)
```

# Rolling Windows

```python
#We can set index in another way
data = pd.read_csv('/home/hasan/DATA SET/germany-electricity-power-for-20062017/opsd_germany_daily.csv',
                  index_col=0, 
                   parse_dates=True)
```

```python
data.head()
```

```python
#mean in daily consumption
columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
daily_mean = data[columns].resample('D').mean()
daily_mean.head()
```

```python
#7D rolling
data_7d = data[columns].rolling(7, center=True).mean()
data_7d.head()
```

```python
fig, ax = plt.subplots()
start,end = '2017-01','2017-06'

ax.plot(weekly_mean.loc[start:end, 'Consumption'], marker='o', color='red', linestyle='-', linewidth=.5, label='weekly consumption')
ax.plot(daily_mean.loc[start:end, 'Consumption'], marker='.', color='green', linestyle='-', linewidth=.5, label='daily consumption')
ax.plot(data_7d.loc[start:end, 'Consumption'], marker='.', color='blue', linestyle='-', linewidth=.5, label='7d rolling')
plt.xlabel('date')
plt.ylabel('value')
plt.legend(loc='best')
plt.title('daily, weekly and 7d rolling values')
```

# Trends

```python
columns = ['Consumption','Wind','Solar','Wind+Solar']
data_365d = data[columns].rolling(window=365, center=True, min_periods=360).mean()
data_365d.head()
```

```python
fig, ax = plt.subplots()
ax.plot(data['Consumption'], marker='.', markersize=2, color='0.6', linestyle=None, label='Daily')
ax.plot(data_7d['Consumption'], linewidth=2, label='7-d Rolling Mean')
ax.plot(data_365d['Consumption'], color='0.2', linewidth=3, label='Trend (365-d Rolling Mean)')
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption')
plt.legend(loc='best')
```

```python
columns = ['Wind','Solar','Wind+Solar']
fig, ax = plt.subplots()
for col in columns:
    ax.plot(data_365d[col], label=col)
    plt.legend(loc='best')
```
